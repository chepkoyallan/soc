# What is order of operations
****
* Have you ever been asked to **evaluate the expression?**  **“simplify”** or **“find the value.”**
* Expressions can be **numerical**, **variable**, where there are symbols or letters in the expression represents _unknown number(s)._
* Now, the **way in which we evaluate** or simplify an expression is known as the **Order of Operations, or Operator Precedence.**
* the rules tell us **which operations or procedures** we are to perform **first** when evaluating a given mathematical expression.
* **“operations”** in mathematics refers to **addition, subtraction, multiplication, division, grouping, or any other mathematical process.**

# Order of operations defined plus _further reading_
***
* **PEMDAS** whereas in Europe you may use **BODMAS**, and in Canada, **BEMDAS or even GEMDAS**
* The acronyms help us to remember which operations receive higher **priority or precedence or rank**
* Math order of operations goes something like this:
    1. **Parentheses or Brackets** represent any Grouping Symbol, and that means you must simplify inside them first
    2. Exponents or Order represents Powers like squaring, cubing, etc.
    3. Multiplication and Division (whichever comes first working left to right)
    4. Addition and Subtraction (whichever comes first working left to right)
