* The table below lists some key words and phrases that are used to describe common mathematical operations.

* In the table below, the letter x is used to represent the unknown. 
* In translation problems, the words sum, total, difference, product and quotient imply at least two parts  
* Use parentheses when a sum or difference is multiplied. For example, the phrase **"the sum of three times a number and five" translates to "3x + 5,"** while the phrase 
* **"Three times the sum of a number and five" translates to "3(x + 5)."**
